```python
import numpy
import scipy

import matplotlib
import matplotlib.pyplot as plot
```

```python
def f(x):
    return numpy.sin(10 * x) + x
```

```python
def df(x):
    return 10 * numpy.cos(10 * x) + 1
```

```python
x_sam = numpy.random.rand(10) + 0.01 * numpy.random.randn(10)
y = f(x_sam)

x_tru = numpy.random.rand(1000)
```

```python
def f_approx(x, a):
    return a * x
```

```python
a_all = numpy.arange(1000)/250. - 1.
```

```python
expected_cost = [numpy.sum([(numpy.sin(10 * x) + x - a * x)**2 for x in x_tru])/1000. for a in a_all]
```

```python
empirical_cost = [numpy.sum([(numpy.sin(10 * x) + x - a * x)**2 for x in x_sam])/10. for a in a_all]
```

```python
matplotlib.rcParams.update({'font.size': 22})
plot.plot(list(a_all), expected_cost, 'b-')
plot.hold('on')
plot.plot(list(a_all), empirical_cost, 'r-')
plot.hold('off')
plot.grid('on')
plot.legend([r'Expected Cost $C$', r'Empirical Cost $\tilde{C}$'])
plot.show()
```
